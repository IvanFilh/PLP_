```python
# Crie um dicionário que é uma agenda e coloque nele os seguintes dados: chave (cpf),
# nome, idade, telefone. O programa deve ler um número indeterminado de dados, criar a
# agenda e imprimir todos os itens do dicionário no formato chave: nome, idade, fone.

agenda = {}

while True:
    cpf = input("Digite o CPF (ou 'sair' para encerrar): ")

    if cpf.lower() == 'sair':
        break

    nome = input("Digite o nome: ")
    idade = input("Digite a idade: ")
    telefone = input("Digite o telefone: ")


    agenda[cpf] = {
        'nome': nome,
        'idade': idade,
        'telefone': telefone
    }


print("Agenda:")
for cpf, dados in agenda.items():
    print(f"{cpf}: {dados['nome']}, {dados['idade']} anos, {dados['telefone']}")

```

```python
# Construa um dicionário com nome, salariobruto, salario liquido,
# para número indeterminado de pessoas, nele será calculado o
# desconto de 8% do INSS e de 10% para IRPF. Preciso que o
# programa seja escrito na Linguagem Python de forma Otimizada..

pessoas = []

while True:
    nome = input("Digite o nome (ou 'sair' para encerrar): ")

    if nome.lower() == 'sair':
        break

    salario_bruto = float(input("Digite o salário bruto: "))


    desconto_inss = salario_bruto * 0.08


    desconto_irpf = salario_bruto * 0.10


    salario_liquido = salario_bruto - desconto_inss - desconto_irpf


    pessoa = {
        'nome': nome,
        'salario_bruto': salario_bruto,
        'salario_liquido': salario_liquido
    }


    pessoas.append(pessoa)


print("\nDados das pessoas:")
for pessoa in pessoas:
    print(f"Nome: {pessoa['nome']}")
    print(f"Salário Bruto: R${pessoa['salario_bruto']:.2f}")
    print(f"Salário Líquido: R${pessoa['salario_liquido']:.2f}")
    print("-" * 30)

```

```python
# Faça um programa em Python para gerar automaticamente numeros entre 0 e 99 de uma cartela de bingo. Sabendo que cada cartela devera conter 5 linhas de 5 numeros, gere estes dados de modo a nao ter numeros repetidos dentro das cartelas. O programa deve exibir na ´
# tela a cartela gerada.

import random


def gerar_cartela():
    cartela = []


    for _ in range(5):
        linha = random.sample(range(100), 5)
        cartela.append(linha)

    return cartela

def exibir_cartela(cartela):
    for linha in cartela:
        print(" ".join(f"{n:2}" for n in linha))

cartela = gerar_cartela()
print("Cartela de Bingo:")
exibir_cartela(cartela)
```

```python
# Faça um Programa em Python que Leia uma matriz 10 x 3 com as notas de 10 alunos em 3 provas. Em seguida, escreva o numero de alunos cuja pior nota foi na prova 1, o número de alunos cuja pior nota foi na prova 2, e o numero de alunos cuja pior nota foi na prova 3. Em caso de empate das piores notas de um aluno, o critério de desempate  e arbitrário, mas o aluno deve ser contabilizado apenas uma vez.


def ler_notas():
    notas = []
    for i in range(10):
        notas_aluno = []
        for j in range(3):
            nota = float(input(f"Digite a nota do aluno {i+1}, prova {j+1}: "))
            notas_aluno.append(nota)
        notas.append(notas_aluno)
    return notas


def pior_nota_aluno(notas):
    pior_notas = []
    for aluno_notas in notas:
        pior_nota = min(aluno_notas)
        pior_notas.append(pior_nota)
    return pior_notas


def contar_piores_notas(notas):
    piores_notas_prova = [0, 0, 0]
    pior_notas_alunos = pior_nota_aluno(notas)

    for i in range(10):
        index_prova = notas[i].index(pior_notas_alunos[i])
        piores_notas_prova[index_prova] += 1

    return piores_notas_prova


notas = ler_notas()

piores_notas = contar_piores_notas(notas)


print(f"Número de alunos cuja pior nota foi na prova 1: {piores_notas[0]}")
print(f"Número de alunos cuja pior nota foi na prova 2: {piores_notas[1]}")
print(f"Número de alunos cuja pior nota foi na prova 3: {piores_notas[2]}")
```

```python
# lê um vetor de 8 posições e, em seguida, lê dois valores X e Y quaisquer correspondentes a duas posições no vetor. O programa então imprime a soma dos valores encontrados nas respectivas posições X e Y.


def ler_vetor():
    vetor = []
    for i in range(8):
        valor = int(input(f"Digite o valor na posição {i}: "))
        vetor.append(valor)
    return vetor

def somar_posicoes(vetor, x, y):
    if x >= 0 and x < len(vetor) and y >= 0 and y < len(vetor):
        soma = vetor[x] + vetor[y]
        return soma
    else:
        return None

vetor = ler_vetor()


x = int(input("Digite o valor de X (0 a 7): "))
y = int(input("Digite o valor de Y (0 a 7): "))

resultado = somar_posicoes(vetor, x, y)

if resultado is not None:
    print(f"A soma dos valores nas posições {x} e {y} é igual a {resultado}.")
else:
    print("As posições informadas estão fora do intervalo válido (0 a 7).")
```

```python
# Faça um programa em Python que receba do usuario dois vetores, ´ A e B, com 10 numeros inteiros cada. Crie um novo vetor denominado C calculando C = A - B. Mostre na tela os dados
# do vetor C.

def ler_vetor():
    vetor = []
    for i in range(10):
        valor = int(input(f"Digite o valor na posição {i}: "))
        vetor.append(valor)
    return vetor


print("Digite os valores do vetor A:")
vetor_A = ler_vetor()

print("\nDigite os valores do vetor B:")
vetor_B = ler_vetor()


vetor_C = [a - b for a, b in zip(vetor_A, vetor_B)]

print("\nVetor C (C = A - B):")
for valor in vetor_C:
    print(valor)
```

```python
# Faça um programa em Python que receba 6 numeros inteiros e mostre: ´
#  • Os numeros pares digitados;
#  • A soma dos numeros pares digitados;
#  • Os numeros   ımpares digitados;
#  • A quantidade de numeros  ımpares


numeros_pares = []
numeros_impares = []


soma_pares = 0


quantidade_impares = 0


for i in range(6):
    numero = int(input(f"Digite o {i+1}º número inteiro: "))

    if numero % 2 == 0:
        numeros_pares.append(numero)
        soma_pares += numero
    else:
        numeros_impares.append(numero)
        quantidade_impares += 1

print("Números pares digitados:")
for num in numeros_pares:
    print(num)


print(f"Soma dos números pares: {soma_pares}")

print("Números ímpares digitados:")
for num in numeros_impares:
    print(num)
print(f"Quantidade de números ímpares: {quantidade_impares}")
```

```python
# Faça um programa em Python que receba do usuario
# um vetor com 10 posições. Em seguida devera ser
# impresso o maior e o menor elemento do vetor.


def encontrar_maior_menor(vetor):
    if len(vetor) == 0:
        return None, None

    maior = menor = vetor[0]

    for elemento in vetor:
        if elemento > maior:
            maior = elemento
        elif elemento < menor:
            menor = elemento

    return maior, menor


def ler_vetor():
    vetor = []
    for i in range(10):
        valor = int(input(f"Digite o valor na posição {i+1}: "))
        vetor.append(valor)
    return vetor


vetor = ler_vetor()


maior, menor = encontrar_maior_menor(vetor)


if maior is not None and menor is not None:
    print(f"O maior elemento do vetor é: {maior}")
    print(f"O menor elemento do vetor é: {menor}")
else:
    print("O vetor está vazio.")
```
